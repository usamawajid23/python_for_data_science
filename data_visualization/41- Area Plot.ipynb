# Area Plot
An area plot, also known as a filled area chart or area graph, is a data visualization that represents quantitative data using colored areas to show the cumulative value of multiple variables over a continuous axis, usually time. It is an extension of the line plot, where the area between the line and the axis is filled with color.

Key features and characteristics of an area plot:

1. Stacked Representation: In an area plot, each data series is stacked on top of the previous one, allowing viewers to observe the total contribution of all variables at each point along the x-axis. The sum of the areas represents the total value at each data point.

2. Time-Series and Continuous Data: Area plots are particularly useful for displaying time-series data or continuous data where the values change smoothly over time or another continuous axis.

3. Emphasis on Trends and Proportions: Area plots are effective for emphasizing the relative proportions and trends of different variables as they contribute to the whole. The colored areas make it easy to identify which variables have the most significant impact on the cumulative total.

4. Identifying Changes and Patterns: Area plots help viewers identify changes and fluctuations in the cumulative values over time. Sudden shifts or trends in the data can be more apparent with the filled area representation.

5. Color Coding: Each area in the plot is typically assigned a distinct color to differentiate between data series. Color choices are essential to ensure clarity and accessibility, especially when there are multiple variables.

6. Limitations: While area plots are useful for displaying cumulative data, they can sometimes be misleading if the values change drastically or overlap significantly. In such cases, it might be challenging to interpret individual values accurately.

Area plots are often used in various fields and scenarios, including:

- Financial Analysis: To visualize the cumulative performance of different investment portfolios over time.

- Population Growth: To represent the cumulative population of multiple cities or countries over several years.

- Project Management: To illustrate the cumulative progress of various project tasks over time.

- Marketing: To display the cumulative sales or market share of different products or brands over specific periods.

Creating effective area plots involves careful consideration of data scaling, color choices, and labeling. It is essential to ensure that the visualization accurately conveys the intended message and insights to the audience. Python libraries like Matplotlib and Seaborn offer functions to create area plots efficiently, enabling data analysts and researchers to communicate complex data patterns with ease.

In Python, the Matplotlib library provides a straightforward way to create area plots to visualize cumulative data. Area plots in Matplotlib are built on top of line plots, and they use the fill_between() function to fill the area between the lines and the x-axis. Here's how you can create area plots in Python using Matplotlib:

Import the necessary libraries:

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import seaborn as sns
````

````python
# Create data
x=range(1,6)
y=[1,4,6,8,4]

# Area plot
plt.fill_between(x, y)
plt.show()
````

````python
# Sample data for demonstration
x = np.arange(1, 11)  # x-axis values
y1 = np.array([2, 4, 6, 8, 7, 5, 3, 6, 8, 9])  # First data series
y2 = np.array([1, 3, 5, 7, 9, 9, 4, 2, 3, 5])  # Second data series
````

````python
plt.figure(figsize=(8, 5))  # Optional: Set the figure size

# Plot the first data series as a line plot
plt.plot(x, y1, label='Data Series 1', color='blue')

# Plot the second data series as a line plot
plt.plot(x, y2, label='Data Series 2', color='green')

# Fill the area between the lines and the x-axis
plt.fill_between(x, y1, y2, color='yellow', alpha=0.5)

# Add labels and a legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Area Plot Example')
plt.legend()

# Show the plot
plt.show()
````

In this example, we first import the required libraries, then prepare some sample data in the form of NumPy arrays. Next, we create the area plot using plt.plot() to draw the line plots for the two data series. The fill_between() function is used to fill the area between the two lines with a specified color (yellow in this case). The alpha parameter adjusts the transparency of the filled area, making it slightly transparent for better visibility.

You can customize the plot further by changing colors, adding more data series, adjusting labels, and exploring other features of Matplotlib.

Area plots are useful for visualizing cumulative data trends and patterns, especially when dealing with time-series or continuous data. They provide a clear representation of how multiple variables contribute to the total over a continuous axis.

````python

# Sample data for demonstration
x = np.arange(1, 11)  # x-axis values
y1 = np.array([2, 4, 6, 8, 7, 5, 3, 6, 8, 9])  # First data series
y2 = np.array([1, 3, 5, 7, 9, 6, 4, 2, 3, 5])  # Second data series

# Create the area plot
plt.fill_between(x, y1, y2, color='skyblue', label='Data Series 1 vs. Data Series 2')

# Add labels and a legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Area Plot Example')
plt.legend()

# Show the plot
plt.show()
````


In this above example, we have the same `x` values and two data series, `y1` and `y2`, which represent the y-values for the two lines we want to plot. Instead of using separate line plots, we directly use `fill_between()` to create the area plot between the two data series. The area between `y1` and `y2` will be filled with the specified color (sky blue in this case).

The `xlabel`, `ylabel`, and `title` functions set the axis labels and the plot title. The `legend()` function adds a legend to the plot, which helps identify the different data series.

This straightforward example showcases the primary purpose of an area plot: to visualize the relationship and cumulative contributions of two data series over a continuous x-axis.

Keep in mind that you can further customize the plot by changing colors, adjusting transparency, modifying labels, and exploring other Matplotlib features. Additionally, area plots are useful for visualizing more complex data scenarios, like multiple data series with different stacking or cumulative trends over time.

**Here's a more detailed example of creating an area plot using Matplotlib in Python with accurate data, including strings and integers:**

````python
# Sample data for demonstration
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct']
sales_product1 = [100, 120, 80, 150, 130, 180, 200, 220, 190, 160]
sales_product2 = [80, 90, 70, 120, 110, 140, 160, 180, 170, 140]

# Create the area plot
plt.fill_between(months, sales_product1, sales_product2, color='green', alpha=0.7, label='Product 1 vs. Product 2')

# Add labels and a legend
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales Comparison between Product 1 and Product 2')
plt.legend()

# Show the plot
plt.show()
````


In this example, we have accurate data for monthly sales of two products, 'Product 1' and 'Product 2', for ten months. The `months` list contains the strings representing the month names, and the `sales_product1` and `sales_product2` lists contain the corresponding integer values for the sales of each product in each month.

The `fill_between()` function is used to create the area plot between the sales of the two products for each month. The `color` parameter sets the color of the filled area, and the `alpha` parameter adjusts the transparency of the filled area.

We use the `xlabel`, `ylabel`, and `title` functions to set the axis labels and the plot title. The `legend()` function adds a legend to the plot, helping to identify the sales of 'Product 1' and 'Product 2' for each month.

This example demonstrates how area plots can be used to compare two variables with both string and integer data. It helps visualize the cumulative sales of the two products over the ten-month period and allows viewers to identify which product performed better during each month.

You can modify the data and labels according to your specific use case to create more personalized area plots.

````python
#  updated example with three additional columns:

# Sample data for demonstration
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct']
sales_product1 = [100, 120, 80, 150, 130, 180, 200, 220, 190, 160]
sales_product2 = [80, 90, 70, 120, 110, 140, 160, 180, 170, 140]
sales_product3 = [70, 80, 90, 100, 110, 120, 130, 140, 150, 160]
expenses_product1 = [50, 60, 55, 70, 65, 80, 90, 95, 75, 65]
expenses_product2 = [40, 45, 50, 55, 60, 65, 70, 75, 80, 85]
expenses_product3 = [30, 35, 40, 45, 50, 55, 60, 65, 70, 75]

# Create the area plots
plt.fill_between(months, sales_product1, sales_product2, color='lightblue', alpha=0.7, label='Product 1 vs. Product 2')
plt.fill_between(months, sales_product2, sales_product3, color='lightgreen', alpha=0.7, label='Product 2 vs. Product 3')
plt.fill_between(months, sales_product3, expenses_product1, color='lightcoral', alpha=0.7, label='Product 3 vs. Expenses Product 1')

# Add labels and a legend
plt.xlabel('Months')
plt.ylabel('Amount')
plt.title('Monthly Sales and Expenses Comparison')
plt.legend()

# Show the plot
plt.show()
````

In this updated example, we added three more columns to the data: `sales_product3`, `expenses_product1`, and `expenses_product2`. Each of these lists contains the corresponding integer values for the sales and expenses of each product in each month.

We used the `fill_between()` function multiple times to create three separate area plots, each representing the comparison between different combinations of products and expenses.

By adding more columns, you can now visualize and compare multiple aspects of the data in the same area plot, providing a comprehensive view of the sales and expenses trends over the ten-month period.

The code of Matplotlib and Pandas in Python to create an area plot with a custom figure size and enable automatic layout for the plot. Here's a step-by-step explanation of each part:

1. `plt.rcParams["figure.figsize"] = [7.50, 3.50]`: This line sets the default figure size for all plots created using Matplotlib. The figure size is specified as a list with two elements representing the width and height of the plot in inches. In this case, the width is set to 7.50 inches, and the height is set to 3.50 inches.

2. `plt.rcParams["figure.autolayout"] = True`: This line enables automatic layout for the plot. When set to `True`, Matplotlib will automatically adjust the figure's layout to fit the content, ensuring that everything is visible within the given figure size.

3. `df = pd.DataFrame(np.random.rand(10, 4), columns=["a", "b", "c", "d"])`: This code creates a Pandas DataFrame named `df` with random data. The DataFrame has 10 rows and 4 columns ("a", "b", "c", and "d"), with each cell containing a random value between 0 and 1 generated by `np.random.rand()`.

4. `df.plot.area()`: This line generates the area plot using the `plot.area()` function of the Pandas DataFrame. Since there are four columns in the DataFrame, four areas will be plotted, stacked on top of each other, showing the cumulative contribution of each column at each data point.

5. `plt.show()`: This command displays the plot on the screen.

In summary, the code creates a 10x4 DataFrame with random data and then plots the data using an area plot. The figure size is customized to have a width of 7.50 inches and a height of 3.50 inches. Additionally, the automatic layout feature ensures that the area plot is properly adjusted and fits within the specified figure size.

`rcParams` stands for "runtime configuration parameters" in Matplotlib. It is a dictionary-like object provided by the Matplotlib library that allows you to customize various aspects of the plotting environment and appearance. With `rcParams`, you can set default values for different plot elements, such as figure size, font sizes, colors, and line styles. These settings are applied to all plots created using Matplotlib in the current session.

When you use `rcParams`, you can adjust the default behavior of Matplotlib without explicitly specifying the settings for each individual plot. This can be especially useful when you want to ensure consistency in the appearance of multiple plots within the same script or notebook.

The `rcParams` dictionary contains a wide range of keys, each corresponding to a specific plot configuration parameter. Some common keys and their meanings include:

- `figure.figsize`: The default figure size, specified as a list with two elements representing the width and height in inches.

- `figure.dpi`: The default resolution of the figure in dots per inch (dpi).

- `figure.autolayout`: A boolean value indicating whether automatic layout adjustment should be applied to figures.

- `font.size`: The default font size for text elements, such as axis labels, tick labels, and legends.

- `axes.grid`: A boolean value indicating whether the grid lines should be shown on the plot.

- `axes.titlelocation`: The default location for the plot title (e.g., "left", "center", "right").

You can modify these settings using the `rcParams` dictionary, as shown in the code snippet you provided earlier:

```python
import matplotlib.pyplot as plt

# Set default figure size and enable automatic layout
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True
```

Here, the `plt.rcParams` dictionary is used to set the default figure size to 7.50 inches width and 3.50 inches height, and automatic layout is enabled for all subsequent plots created with Matplotlib.

Remember that `rcParams` is a global setting, and any changes you make to it will affect all the plots in the current session unless you explicitly reset or modify the parameters again.

````python
# Example referance: "https://python-graph-gallery.com/area-fill-between-two-lines-in-matplotlib/" 

time = np.arange(12)
income = np.array([5, 9, 6, 6, 10, 7, 6, 4, 4, 5, 6, 4])
expenses = np.array([6, 6, 8, 3, 6, 9, 7, 8, 6, 6, 4, 8])

# Initialize figure and axis
fig, ax = plt.subplots(figsize=(8, 8))

# Plot lines
ax.plot(time, income, color="green")
ax.plot(time, expenses, color="red")

# Fill area when income > expenses with green
ax.fill_between(
    time, income, expenses, where=(income > expenses), 
    interpolate=True, color="green", alpha=0.25, 
    label="Positive"
)

# Fill area when income <= expenses with red
ax.fill_between(
    time, income, expenses, where=(income <= expenses), 
    interpolate=True, color="red", alpha=0.25,
    label="Negative"
)

ax.legend();
````
