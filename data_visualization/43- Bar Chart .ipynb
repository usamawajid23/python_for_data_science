# Bar Chart

A bar chart is a data visualization used to represent categorical data in the form of rectangular bars. It displays the frequency or value of each category in a dataset, making it easy to compare different categories visually. Bar charts are particularly useful for showing discrete data and identifying trends or comparisons between categories.

Key features and characteristics of a bar chart:

1. Categories: The x-axis of the bar chart represents the different categories or groups of data, and the y-axis shows the corresponding values or counts.

2. Bars: Each category is represented by a rectangular bar whose length or height is proportional to the value it represents.

3. Discrete Data: Bar charts are most suitable for displaying data with distinct categories, such as product names, months, cities, or survey responses.

4. Ordering: The order of categories on the x-axis can be customized to emphasize certain patterns or trends.

5. Width and Spacing: The width and spacing between bars can be adjusted to control the visual appearance and readability of the chart.

6. Horizontal and Vertical Bar Charts: Bar charts can be oriented either horizontally or vertically, depending on the preference and data presentation requirements.

Bar charts are commonly used in various scenarios, including:

- Comparing sales figures of different products or services.
- Visualizing survey responses for multiple choices or options.
- Analyzing the distribution of population in different cities or regions.
- Displaying the performance of students across various subjects.

Creating a bar chart in Python using Matplotlib is straightforward. You can use the `plt.bar()` function to plot the bars and customize the appearance of the chart with labels, colors, and other aesthetics.


### Types of bar charts

There are several types of bar charts, each serving a different purpose and suitable for different types of data visualization. The main types of bar charts are as follows:

1. Vertical Bar Chart: The most common type of bar chart, where bars are represented vertically along the x-axis. Each bar's height represents the value or frequency of a category or data point.

2. Horizontal Bar Chart: Similar to the vertical bar chart, but the bars are represented horizontally along the y-axis. This orientation is useful when you have long category names or want to emphasize comparisons between categories.

3. Stacked Bar Chart: In a stacked bar chart, multiple data series are stacked on top of each other to show the cumulative total of each category. This chart is useful for comparing the total values of multiple datasets for each category.

4. Grouped Bar Chart: In a grouped bar chart, multiple data series are grouped side by side for each category. This chart allows for easy comparison of values between different datasets within the same category.

5. 100% Stacked Bar Chart: This is a variation of the stacked bar chart, where each stacked bar's height represents the percentage of the whole for each category. All the bars have the same height, representing 100% of the data for each category.

6. Clustered Bar Chart: A combination of a grouped bar chart and a stacked bar chart, where multiple datasets are grouped side by side, and each group is further divided into stacked bars to show the contribution of different categories within each dataset.

7. Floating Bar Chart: In this chart, bars are floating from a baseline instead of starting at the x-axis or y-axis. It is often used to visualize ranges or timelines.

Each type of bar chart has its unique advantages and use cases. Choosing the appropriate type of bar chart depends on the nature of the data you want to represent and the specific insights you want to convey. It is essential to consider the context of your data and the audience when selecting the right type of bar chart for effective data visualization.

Here's a simple example of creating a vertical bar chart

```python
import matplotlib.pyplot as plt

# Sample data for a bar chart
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 30, 55]

# Create the bar chart
plt.bar(categories, values, color='blue')

# Add labels and a title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Vertical Bar Chart Example')

# Show the plot
plt.show()
```

In this example, we have a list of categories and their corresponding values. The `plt.bar()` function is used to create a vertical bar chart, with the categories on the x-axis and the values on the y-axis. The bars are displayed in blue color by setting `color='blue'`. You can modify various aspects of the bar chart, such as bar colors, bar width, and axis labels, to customize the visualization according to your specific requirements.

### Horizontal Bar chart


Let's assume we want to create a bar chart to represent the number of books read by five students during a particular month.

```python
import matplotlib.pyplot as plt

# Sample data for the number of books read by each student
students = ['Alice', 'Bob', 'Charlie', 'David', 'Eve']
books_read = [10, 15, 8, 12, 7]

# Create the bar horizontal chart
plt.barh(students, books_read, color='purple')

# Add labels and a title
plt.xlabel('Student')
plt.ylabel('Number of Books Read')
plt.title('Number of Books Read by Students in a Month')

# Show the plot
plt.show()
```


In this example, we have a list `students` containing the names of five students, and another list `books_read` containing the corresponding number of books each student read during the month.

The `plt.bar()` function creates a vertical bar chart with the students' names on the x-axis and the number of books read on the y-axis. The bars are colored in purple.

You can use this sample data to create a bar chart for any similar dataset you have, such as the number of sales by different salespersons, the scores of students in a test, the sales of various products, or any other discrete data that you wish to visualize. Simply replace the `students` and `books_read` lists with your data, and the bar chart will adapt accordingly.

### Stacked Bar Chart:

```python
# Sample data for a stacked bar chart
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values1 = [20, 15, 25, 30]
values2 = [5, 25, 5, 25]

# Create the stacked bar chart
plt.bar(categories, values1, color='blue', label='Value 1')
plt.bar(categories, values2, bottom=values1, color='green', label='Value 2')

# Add labels and a title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart Example')
plt.legend()

# Show the plot
plt.show()
```

```python
N = 5
 
boys = (20, 35, 30, 35, 27)
girls = (25, 32, 34, 20, 25)
boyStd = (2, 3, 4, 1, 2)
girlStd = (3, 5, 2, 3, 3)
ind = np.arange(N)  
width = 0.35 
 
fig = plt.subplots(figsize =(10, 7))
p1 = plt.bar(ind, boys, width, yerr = boyStd)
p2 = plt.bar(ind, girls, width,
             bottom = boys, yerr = girlStd)
 
plt.ylabel('Contribution')
plt.title('Contribution by the teams')
plt.xticks(ind, ('T1', 'T2', 'T3', 'T4', 'T5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('boys', 'girls'))
 
plt.show()
```

```python
 
# create data
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])
 
# plot bars in stack manner
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.bar(x, y3, bottom=y1+y2, color='y')
plt.bar(x, y4, bottom=y1+y2+y3, color='g')
plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.title("Scores by Teams in 4 Rounds")
plt.show()
```

### Grouped bar chart

```python
# Sample data for a grouped bar chart
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values1 = [20, 15, 25, 30]
values2 = [10, 20, 30, 40]

# Create the grouped bar chart
bar_width = 0.35
x_pos = range(len(categories))
plt.bar(x_pos, values1, color='blue', width=bar_width, label='Value 1')
plt.bar([pos + bar_width for pos in x_pos], values2, color='green', width=bar_width, label='Value 2')

# Add labels and a title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Grouped Bar Chart Example')
plt.xticks([pos + bar_width / 2 for pos in x_pos], categories)
plt.legend()

# Show the plot
plt.show()
```

100% stacked bar chart

```python
# Sample data for a 100% stacked bar chart
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values1 = [20, 15, 25, 30]
values2 = [10, 20, 30, 40]

# Calculate the percentage for each category
total = [v1 + v2 for v1, v2 in zip(values1, values2)]
percentage1 = [v1 / t * 100 for v1, t in zip(values1, total)]
percentage2 = [v2 / t * 100 for v2, t in zip(values2, total)]

# Create the 100% stacked bar chart
plt.bar(categories, percentage1, color='blue', label='Value 1')
plt.bar(categories, percentage2, bottom=percentage1, color='green', label='Value 2')

# Add labels and a title
plt.xlabel('Categories')
plt.ylabel('Percentage')
plt.title('100% Stacked Bar Chart Example')
plt.legend()

# Show the plot
plt.show()
```

### clustered bar chart

```python
import pandas as pd
df = pd.read_clipboard()
df
```

```python
# plot a Stacked Bar Chart using matplotlib
df.plot(
    x = 'Name',
    kind = 'barh',
    stacked = True,
    title = 'Stacked Bar Graph',
    mark_right = True)
```

```python
# plot a Stacked Bar Chart using matplotlib
df.plot(
  x = 'Name', 
  kind = 'barh', 
  stacked = True, 
  title = 'Percentage Stacked Bar Graph', 
  mark_right = True)
  
df_total = df["Studied"] + df["Slept"] + df["Other"]
df_rel = df[df.columns[1:]].div(df_total, 0)*100
  
for n in df_rel:
    for i, (cs, ab, pc) in enumerate(zip(df.iloc[:, 1:].cumsum(1)[n], 
                                         df[n], df_rel[n])):
        plt.text(cs - ab / 2, i, str(np.round(pc, 1)) + '%', 
                 va = 'center', ha = 'center')
```

```python
# Sample data for a clustered bar chart
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values1 = [20, 15, 35, 30]
values2 = [10, 20, 30, 20]

# Create the clustered bar chart
bar_width = 0.35
x_pos = range(len(categories))
plt.bar(x_pos, values1, color='blue', width=bar_width, label='Value 1')
plt.bar([pos for pos in x_pos], values2, color='green', width=bar_width, label='Value 2')

# Add labels and a title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Clustered Bar Chart Example')
plt.xticks(x_pos, categories)
plt.legend()

# Show the plot
plt.show()
```

### Back-to-Back Bar Chart:
    Back-to-Back Bar Chart is just a combination of two bar charts drawn with respect to a axis.

Example 1: Simple Back-to-Back Bar Chart

```python
# create data
A = np.array([3,6,9,4,2,5])
X = np.arange(6)
 
# plot the bars
plt.bar(X, A, color = 'r')
plt.bar(X, -A, color = 'b')
plt.title("Back-to-Back Bar Chart")
plt.show()
```

```python
# create data
A = np.array([3,6,9,4,2,5])
B = np.array([2,8,1,9,7,3])
X = np.arange(6)

# plot the bars
plt.barh(X, A, color = 'r')
plt.barh(X, -B, color = 'b')
plt.title("Back-to-Back Bar Chart")
plt.show()
```

```python
# create data
A = np.array([3,6,9,4,2,5])
B = np.array([2,8,1,9,7,3])
X = np.arange(6)

# plot the bars
plt.barh(X, A, align='center',
		alpha=0.9, color = 'y')

plt.barh(X, -B, align='center',
	 	alpha=0.6, color = 'c')

plt.grid()
plt.title("Back-to-Back Bar Chart")
plt.ylabel("Indexes")
plt.xlabel("Values")
plt.show()
```
