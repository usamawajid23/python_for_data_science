# Histogram plot | Data Visualization 

A histogram plot is a graphical representation used to visualize the distribution of a dataset. It provides a visual summary of the frequency or count of data points that fall within specific intervals or "bins" along the range of the data. Histograms are commonly used for exploring the shape and spread of continuous or numerical data.

Key features and characteristics of a histogram plot:

1. Bins: The data range is divided into several non-overlapping intervals or bins. The width of each bin is determined based on the data's range and the desired level of granularity.

2. Frequency: The height of each bar in the histogram represents the frequency or count of data points falling within the corresponding bin. The higher the bar, the more data points are present in that particular interval.

3. Continuous Data: Histograms are most appropriate for continuous data, such as measurements on a scale, temperatures, test scores, etc., as opposed to discrete data, like categories or counts.

4. Skewness: The shape of the histogram can reveal important information about the distribution of the data. For example, if the histogram is skewed to the left or right, it indicates an asymmetrical distribution.

5. Central Tendency: The central peak in the histogram often represents the central tendency or mode of the data, which can provide insights into the most typical values.

6. Outliers: Outliers (extreme values) can be visually identified as data points that fall far outside the main body of the histogram.

Histogram plots are widely used in various fields, including:

- Analyzing exam scores or grades to understand the performance of students.
- Examining income distribution to assess economic disparities.
- Investigating the distribution of temperatures or rainfall amounts in climatology.
- Studying the frequency of customer purchases to understand buying patterns.

Creating a histogram plot in Python is straightforward. You can use libraries like Matplotlib or Seaborn to generate histograms from your data, specifying the number of bins and customizing other plot aesthetics to enhance visualization. Histograms are valuable for gaining insights into data distributions, identifying outliers, and making data-driven decisions.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from matplotlib import style
```

```python
age_men = 25,11,68,18,27,28,15,43,58,63,43,65,51,36,33,26,23,35,49,58
```

```python
plt.hist(age_men)
```

```python
# after created bins = 5
plt.hist(age_men, bins=5)
```

```python
bins =[10,20,30,40,50,60,70]
plt.hist(age_men, bins=bins, edgecolor='black')
```

```python
bins =[10,30,40,70]
plt.hist(age_men, bins=bins, edgecolor='black')
```

```python
#adding style to the plot
style.use('ggplot')
#plotting the histogram
plt.hist(age_men, bins=10, edgecolor='black', color='blue', 
         rwidth=0.6)  # bar width
#plotting the xlabel
plt.xlabel('age of people')
#plotting the ylabel
plt.ylabel('number of people')
#plotting the title of the graph
plt.title('Age vs number of people')
```

```python
age_men = 25,11,68,18,27,28,15,43,58,63,43,65,51,36,33,26,23,35,49,58
age_women = 48,57,59,25,19,37,18,56,22,25,56,25,14,49,53,45,46,19,28,70
```

```python
plt.hist([age_men, age_women])
```

```python
plt.hist([age_men, age_women], bins=10, color=['blue', 'orange'],rwidth=0.9, label=['men', 'women'])
plt.legend(loc='upper right')
```

```python
#plot the histogram
plt.hist([age_men, age_women], bins=10, 
         color=['blue', 'orange'],  # color of bar
         histtype='barstacked',   # barstacked is a bar-type histogram where multiple data are stacked on top of each other
         rwidth=0.9,              # width of bar
         label=['men', 'women'])  # lable shown color on the top right corner

plt.legend(loc='upper right')
```

```python
#plot the histogram
plt.hist([age_men, age_women], bins=10, 
         color=['blue', 'orange'],
         rwidth=0.9, histtype='barstacked', label=['men', 'women'])
#add xlabel, ylabel and title
plt.xlabel('age of people')
plt.ylabel('number of people')
plt.title('Age vs number of people')
plt.legend(loc='upper right')
```

```python
#set the figure size
plt.figure(figsize=(10,7))

#add style to the plot
style.use('ggplot')

#plot the histogram
plt.hist([age_men, age_women], bins=10, 
         color=['blue', 'orange'], histtype='barstacked',
         rwidth=0.9, label=['men', 'women'])

#add xlabel, ylabel and title
plt.xlabel('age of people')
plt.ylabel('number of people')
plt.title('Age vs number of people')
plt.legend(loc='upper right')
```

```python
style.use('ggplot')
plt.figure(figsize=(10,7))

plt.hist(age_men,
         bins=10,
         color='blue',
         rwidth=0.8,
         label='Men')

plt.hist(age_women,
         bins=10,
         color='orange',
         rwidth=0.8,
         label = 'Women')

plt.xlabel('Age of people')
plt.ylabel('No. of people')
plt.title('Age vs number of people')
plt.legend(loc='upper right')
```

```python
style.use('ggplot')
plt.figure(figsize=(10,7))

plt.hist(age_men,
         bins=10,
         color='blue',
         histtype='step',   #histogram plot type
         rwidth=0.8,
         label='Men')

plt.hist(age_women,
         bins=10,
         color='orange',
         histtype='step',
         rwidth=0.8,
         label = 'Women')

plt.xlabel('Age of people')
plt.ylabel('No. of people')
plt.title('Age vs number of people')
plt.legend(loc='upper right')
```

```python
style.use('ggplot')
plt.figure(figsize=(15,5))

plt.subplot(1,2,2)

plt.hist(age_men, bins=10, color='blue',
         histtype='bar', rwidth=0.8)
plt.xlabel('age of men')
plt.ylabel('No of people')
plt.title('Age vs number of people')

plt.subplot(1,2,1)

plt.hist(age_women, bins=10, color='blue',
         histtype='bar', rwidth=0.8)
plt.xlabel('age of women')
plt.ylabel('No of people')
plt.title('Age vs number of people')
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.hist(age_men, bins=10, color='blue', histtype='bar', rwidth=0.8)
plt.xlabel('age of men')
plt.ylabel('No of people')
plt.title('Age vs number of people')
plt.subplot(2,2,4)
plt.hist(age_women, bins=10, color='blue', histtype='bar', rwidth=0.8)
plt.xlabel('age of women')
plt.ylabel('No of people')
plt.title('Age vs number of people')
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)

mu_x = 200
sigma_x = 25
x = np.random.normal(mu_x, sigma_x, size=100)

mu_w = 200
sigma_w = 10
w = np.random.normal(mu_w, sigma_w, size=100)

fig, axs = plt.subplots(nrows=2, ncols=2)

axs[0, 0].hist(x, 20, density=True, histtype='stepfilled', facecolor='g',
               alpha=0.75)
axs[0, 0].set_title('stepfilled')

axs[0, 1].hist(x, 20, density=True, histtype='step', facecolor='g',
               alpha=0.75)
axs[0, 1].set_title('step')

axs[1, 0].hist(x, density=True, histtype='barstacked', rwidth=0.8)
axs[1, 0].hist(w, density=True, histtype='barstacked', rwidth=0.8)
axs[1, 0].set_title('barstacked')

# Create a histogram by providing the bin edges (unequally spaced).
bins = [100, 150, 180, 195, 205, 220, 250, 300]
axs[1, 1].hist(x, bins, density=True, histtype='bar', rwidth=0.8)
axs[1, 1].set_title('bar, unequal bins')

fig.tight_layout()
plt.show()
```
