# Pie Chart - Data Visualization 
A pie chart is a circular data visualization used to represent the proportions of different categories in a dataset. The entire pie represents the whole data, and each "slice" of the pie corresponds to a specific category's proportion, typically expressed as a percentage of the whole. Pie charts are ideal for displaying data with a small number of categories and when you want to emphasize the relationship of each part to the whole.

Key features and characteristics of a pie chart:

1. Slices: Each slice of the pie represents a category or data point, and its size is proportional to the percentage it contributes to the total.

2. Proportions: The central angle of each slice is determined by the proportion of the corresponding category's data in the whole dataset.

3. 100% Pie Chart: In a 100% pie chart, the sum of all slices represents 100% of the data. This type of pie chart is useful when comparing the relative proportions of different categories within the whole.

4. Data Labels: Data labels are often added to each slice to display the actual percentage or value represented by that slice.

5. Exploded Pie Chart: You can "explode" or separate one or more slices from the center of the pie to highlight a specific category or emphasize its importance.

Creating a pie chart in Python using Matplotlib is straightforward. You can use the `plt.pie()` function to plot the pie chart, specifying the data and labels.

Here's an example of creating a simple pie chart:

````python
import matplotlib.pyplot as plt

# Sample data for the pie chart
categories = ['Category A', 'Category B', 'Category C', 'Category D']
proportions = [30, 25, 15, 30]  # Represents 30%, 25%, 15%, and 30% respectively

# Create the pie chart
plt.pie(proportions, labels=categories, autopct='%1.1f%%', startangle=90, colors=['blue', 'green', 'orange', 'red'])

# Add a title
plt.title('Proportions of Categories in the Dataset')

# Show the plot
plt.show()
````


In this example, we have a list `categories` representing the different categories, and another list `proportions` representing the proportions or percentages of each category in the whole dataset. The `plt.pie()` function creates the pie chart with the corresponding proportions. The `labels` parameter specifies the category labels, `autopct` adds percentage labels on each slice, and `startangle` determines the angle at which the first slice starts. The `colors` parameter allows customizing the colors of each slice.

Pie charts are a useful visualization for representing categorical data when you want to quickly understand the distribution of categories and their relative proportions. However, it's essential to be cautious when using pie charts, especially with datasets containing many categories or where the differences in proportions are subtle, as they can sometimes be less effective than other types of charts in conveying information accurately.

The parameter `startangle` in a pie chart determines the starting angle or rotation of the first slice in the chart. It specifies the angle at which the pie chart starts drawing the first slice. The slices of the pie are drawn in a clockwise direction from the starting angle.

In the code snippet you provided:

```python
data.plot(kind='pie', autopct="%0.1f%%", labels=classes, startangle=45)
```

The `startangle=45` sets the starting angle of the first slice to 45 degrees. This means that the first slice of the pie chart will begin at a 45-degree angle measured in a clockwise direction from the 12 o'clock position. The subsequent slices will follow in the clockwise direction from there.

Setting the `startangle` can be useful to highlight a specific category or to adjust the orientation of the pie chart to better fit the visual representation of the data. By default, the starting angle is set to 0 degrees, which places the first slice on the right side of the pie chart.

If you change the `startangle` value to a different angle, the appearance of the pie chart will change accordingly, but the proportions and data represented by the slices will remain the same. For example, using `startangle=90` would rotate the pie chart by 90 degrees counterclockwise, placing the first slice on the bottom side of the pie chart.

````python
data = df['Species'].value_counts()
classes = ['Iris-setosa','Iris-Versicolor','Iris-Virginica']
data.plot(kind='pie', autopct="%0.1f%%", labels = classes, startangle=45 )
````

````python
fig = plt.figure(figsize=(10,10))
# add colors
colors = ("#ffc100", "#ff7400","crimson")

explode = (0,0.1,0)

data = df['Species'].value_counts()
classes = ['Iris-setosa','Iris-Versicolor','Iris-Virginica']
data.plot(kind='pie', autopct="%0.1f%%", labels = classes, startangle=45, colors=colors, explode = explode) #add explode here
````

In this code, the `explode` variable is defined as a tuple containing three values: `(0.1, 0.1, 0.1)`.

The `explode` variable is used in the context of a pie chart, and it determines the degree of separation (or "explosion") of individual slices from the center of the pie.

Here's what each value in the `explode` tuple represents:

1. `0.1`: The first value (0.1) represents the degree of separation for the first slice of the pie chart. It means that the first slice will be slightly separated from the center of the pie.

2. `0.1`: The second value (0.1) represents the degree of separation for the second slice of the pie chart. It means that the second slice will also be slightly separated from the center of the pie.

3. `0.1`: The third value (0.1) represents the degree of separation for the third slice of the pie chart. It means that the third slice will also be slightly separated from the center of the pie.

The values in the `explode` tuple are relative to the radius of the pie chart. A value of 0.1 means that the slice will be separated by 10% of the pie's radius from the center. Similarly, a value of 0.2 would mean a separation of 20%, and so on.

Using the `explode` parameter allows you to visually highlight specific slices in a pie chart by pulling them away from the center. This can be useful when you want to emphasize a particular category or data point.

For example, in a pie chart representing the distribution of different expenses, you could use `explode` to highlight the largest expense by pulling its slice slightly away from the center, making it stand out more prominently.

````python
fig = plt.figure(figsize=(10,10))
# add colors
colors = ("#ffc100", "#ff7400","crimson")

explode = (0.1,0.1,0.1)

data = df['Species'].value_counts()
classes = ['Iris-setosa','Iris-Versicolor','Iris-Virginica']
data.plot(kind='pie', autopct="%0.1f%%", labels = classes, startangle=45, colors=colors, explode = explode) #add explode here
````

In the context of the code provided, "wp" stands for **"wedge properties."** The variable name "wp" is used as a shorthand to represent a dictionary that specifies the properties of the wedges (slices) in the pie chart.

In the code snippet:

```python
wp = {'linewidth': 2, 'edgecolor': "black"}
```

When you pass the "wp" dictionary as the "wedgeprops" parameter to the `plt.pie()` function, it applies the specified properties to the edges of the pie chart slices, affecting their appearance. This allows you to customize the appearance of the pie chart and make it visually appealing and informative.

In this code, the variable `wp` is defined as a dictionary with two key-value pairs:

1. `'linewidth': 2`: This key-value pair sets the width of the edges (lines) of the pie chart wedges (slices) to 2 points. The `linewidth` parameter controls the thickness of the lines outlining each slice of the pie chart.

2. `'edgecolor': "black"`: This key-value pair sets the color of the edges (lines) of the pie chart wedges to black. The `edgecolor` parameter allows you to customize the color of the lines outlining each slice.

These parameters are used in the context of a pie chart, and they affect the appearance of the individual slices of the pie chart.

When you plot the pie chart using `plt.pie()` and pass `wedgeprops=wp` as a parameter, the specified properties in the `wp` dictionary will be applied to the edges of the pie chart slices. In this case, the edges of the slices will have a width of 2 points and will be colored black.

Customizing the properties of the pie chart slices, such as linewidth and edgecolor, can help enhance the visual appeal and readability of the chart, making it easier for viewers to interpret the data effectively.

````python
fig = plt.figure(figsize=(10,10))
# add colors
colors = ("#ffc100", "#ff7400","crimson")
wp = {'linewidth':1, 'edgecolor':"black"} # add wedge propertie

explode = (0.1,0.1,0.1)

data = df['Species'].value_counts()
classes = ['Iris-setosa','Iris-Versicolor','Iris-Virginica']
data.plot(kind='pie', autopct="%0.1f%%", labels = classes, startangle=45, colors=colors,
          explode = explode,
          shadow = True,     # add shadow
          wedgeprops = wp)  # add wedge properties
````

### Step-by-step explanation of the full code:

1. `fig = plt.figure(figsize=(10,10))`: This line creates a new figure (canvas) for the plot with a size of 10 inches in width and 10 inches in height. The figure will be used to display the pie chart.

2. `colors = ("#ffc100", "#ff7400", "crimson")`: This line defines a tuple `colors` containing three color codes representing the colors for the slices of the pie chart. The first two colors are custom shades of orange, and the third color is "crimson."

3. `wp = {'linewidth':1, 'edgecolor':"black"}`: This line defines a dictionary `wp` that specifies the properties of the wedges (slices) in the pie chart. It sets the linewidth of the wedge edges to 1 point and the edge color to black.

4. `textprops = {'fontstyle':'italic', 'fontweight':'heavy'}`: This line defines a dictionary `textprops` that specifies the properties of the text labels inside the pie chart. It sets the font style to italic and the font weight to heavy.

5. `explode = (0.1, 0.1, 0.1)`: This line defines a tuple `explode` that determines the degree of separation (or "explosion") of individual slices from the center of the pie chart. Each value in the tuple represents the separation for the corresponding slice. In this case, all three slices will be slightly separated.

6. `data = df['Species'].value_counts()`: This line calculates the count of each unique species in the "Species" column of the DataFrame `df`. The result is a Pandas Series containing the counts of each species.

7. `classes = ['Iris-setosa', 'Iris-Versicolor', 'Iris-Virginica']`: This line creates a list `classes` containing the names of the three species in the "Iris" dataset. It provides labels for each slice of the pie chart.

8. `data.plot(kind='pie', autopct="%0.1f%%", labels=classes, startangle=45, colors=colors, explode=explode, shadow=True, wedgeprops=wp, textprops=textprops)`: This line creates the pie chart using the data from the `data` Series. The `kind='pie'` parameter specifies that we want to create a pie chart. The `autopct="%0.1f%%"` adds percentage labels on each slice with one decimal point, displaying the proportion of each species in the chart. The `labels=classes` parameter sets the labels for each slice using the `classes` list. The `startangle=45` specifies the angle at which the first slice starts. The `colors=colors` parameter sets the colors of the slices. The `explode=explode` parameter applies the separation defined in the `explode` tuple. The `shadow=True` adds a shadow effect to the pie chart. The `wedgeprops=wp` sets the properties of the wedges, including the linewidth and edge color. The `textprops=textprops` sets the properties of the text labels inside the pie chart.

9. `plt.legend(title="species")`: This line adds a legend to the pie chart, with the title "species." The legend shows the labels of the different slices (species names) along with their corresponding colors.

10. `plt.title("Different iris species", loc='center', color='black', fontsize='19', fontweight='bold')`: This line sets the title of the pie chart to "Different iris species." The `loc='center'` parameter centers the title. The `color='black'` sets the title color to black, `fontsize='19'` sets the font size to 19, and `fontweight='bold'` sets the font weight to bold.

11. `plt.show()`: This line displays the pie chart on the screen. The chart will show the proportions of different iris species in the dataset, with custom colors, separation, and text styles for better visualization and readability.

````python
fig = plt.figure(figsize=(10,10))

colors = ("#ffc100", "#ff7400","crimson")
wp = {'linewidth':1, 'edgecolor':"black"}
textprops = {'fontstyle':'italic','fontweight':'heavy'} # add textprops 

explode = (0.1,0.1,0.1)

data = df['Species'].value_counts()
classes = ['Iris-setosa','Iris-Versicolor','Iris-Virginica']
data.plot(kind='pie', autopct="%0.1f%%", labels = classes, startangle=45, colors=colors,
          explode = explode,
          shadow = True,     
          wedgeprops = wp,
          textprops = textprops)  # add here
plt.legend(title="species")       # title
plt.title("Different iris species", loc='center', color='black', fontsize='19', fontweight='bold')
plt.show()
````
