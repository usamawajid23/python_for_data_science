## Line Plot

A line plot, also known as a line chart or line graph, is a data visualization that represents data points connected by straight lines. It is commonly used to show trends and changes over time or the relationship between two continuous variables.

The x-axis represents the independent variable, while the y-axis displays the dependent variable. Line plots are effective for visualizing time-series data, scientific experiments, business metrics, and social science research, offering a clear and intuitive way to identify patterns and fluctuations in the data.

## When to use line plot
Line plots are suitable for various scenarios when you want to visualize and analyze trends, patterns, or relationships in your data. Here are some specific situations when it's appropriate to use a line plot:

1. Time-Series Data: Line plots are commonly used to display data collected over time, such as stock prices, temperature changes, population growth, or monthly sales figures. They help reveal trends, seasonal patterns, and long-term fluctuations.

2. Continuous Data Trends: When you have continuous data points and want to understand how a variable changes in response to another variable (e.g., temperature vs. pressure in an experiment), a line plot effectively illustrates the relationship between the two variables.

3. Tracking Metrics: Line plots are ideal for tracking and comparing metrics over time, like website traffic, revenue, or social media engagement. They provide a clear view of performance changes and help identify growth or decline.

4. Comparative Analysis: Line plots can be used to compare trends between multiple categories or groups. For example, you can compare the sales performance of different products over a specific period.

5. Identifying Anomalies: Line plots can help detect anomalies in data, such as sudden spikes or drops in values, which may require further investigation.

6. Showing Growth or Decline: When you want to illustrate the progression of a variable, such as the growth of a company's profits or the decline in pollution levels over the years, line plots provide a concise representation.

7. Correlation Analysis: If you suspect a correlation between two variables, a line plot can provide a quick visual assessment of whether they move in a similar or opposite direction.

8. Communicating Trends Clearly: Line plots are easy to interpret and widely understood, making them a suitable choice for presenting data to a broad audience, including stakeholders, clients, or the general public.

Remember, while line plots are effective for showing trends and patterns over time or between continuous variables, they may not be suitable for categorical data or discrete data points, where other types of visualizations like bar charts or scatter plots may be more appropriate. The choice of visualization depends on the nature of the data and the specific insights you want to communicate.

#### Line with asymptote
An asymptote can be added to the plot. To do that, use plt.annotate(). There’s lso a dotted line in the plot below. You can play around with the code to see how it works.


```python


x = np.linspace(-1, 1, 50)
y1 = 2*x + 1
y2 = 2**x + 1

plt.figure(figsize=(12, 8))  
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')

ax = plt.gca()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))


x0 = 1
y0 = 2*x0 + 1

plt.scatter(x0, y0, s = 66, color = 'b')
plt.plot([x0, x0], [y0, 0], 'k-.', lw= 2.5)

plt.annotate(r'$2x+1=%s$' % 
             y0, 
             xy=(x0, y0), 
             xycoords='data',
             
             xytext=(+30, -30),
             textcoords='offset points',
             fontsize=16,  
             arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.2')
            )

plt.text(0, 3, 
         r'$This\ is\ a\ good\ idea.\ \mu\ \sigma_i\ \alpha_t$',
         fontdict={'size':16,'color':'r'})

plt.show()
```

#### Line with text scale
It doesn’t have to be a numeric scale. The scale can also contain textual words like the example below. In plt.yticks() we just pass a list with text values. These values are then show against the y axis.


```python
x = np.linspace(-1, 1, 50)
y1 = 2*x + 1
y2 = 2**x + 1

plt.figure(num = 3, figsize=(8, 5))  
plt.plot(x, y2)

plt.plot(x, y1, 
         color='red',  
         linewidth=1.0,  
         linestyle='--'  
        )

plt.xlim((-1, 2))  
plt.ylim((1, 3))  

new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)
plt.yticks([-2, -1.8, -1, 1.22, 3],
          [r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$readly\ good$'])

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

plt.show()
```
